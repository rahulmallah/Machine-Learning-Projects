```python
import transformers
from transformers import pipeline
import gradio as gr
```

```python
pipe=pipeline(task='text2text-generation',model='facebook/m2m100_418M')
```

```python
result=pipe('I will be a tensorflow developer',forced_bos_token_id=pipe.tokenizer.get_lang_id(lang='hi'))
```

```python
result
```

```python
result[0]['generated_text']
```

```python
def translate_gradio(input):
    result=pipe(input,forced_bos_token_id=pipe.tokenizer.get_lang_id(lang='hi'))
    return result[0]['generated_text']
```

```python
translate_interface = gr.Interface(fn = translate_gradio,
                                   inputs="text",
                                   outputs="text")
```

```python
translate_interface.launch(share=True)
```
