```python
import pandas as pd
```

```python
train_df=pd.read_csv('../input/our-data/dataset/train.csv')
test_df=pd.read_csv('../input/our-data/dataset/test.csv')
sub_df=pd.read_csv('../input/our-data/dataset/sample_submission.csv')
```

```python
train_df.drop(columns=['Product_id','Stall_no','instock_date','Customer_name'],axis=1,inplace=True)
test_df.drop(columns=['Product_id','Stall_no','instock_date','Customer_name'],axis=1,inplace=True)
```

```python
def encoder(data):
    cat_features=data.select_dtypes(include='object').columns
    for col in cat_features:
        z=pd.get_dummies(data[col])
        data.drop(columns=col,axis=1,inplace=True)
        data=pd.merge(z,data,left_index=True,right_index=True)
        
    return data
```

```python
pd.set_option("display.max_columns",None)
```

```python
train_df=encoder(train_df)
train_df
```

```python
test_df=encoder(test_df)
test_df
```

```python
train_df.isnull().sum()
```

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
imputer = IterativeImputer(max_iter = 8, random_state = 0, skip_complete = True, n_nearest_features = 12)
```

```python
train_df_impute=imputer.fit_transform(train_df)
train_df=pd.DataFrame(train_df_impute,columns=train_df.columns)


test_df_impute=imputer.fit_transform(test_df)
test_df=pd.DataFrame(test_df_impute,columns=test_df.columns)
```

```python
train_df.isnull().sum()
```

```python
X=train_df.drop(columns='Selling_Price',axis=1)
y=train_df['Selling_Price']
```

```python
X
```

```python
y
```

```python
from sklearn.preprocessing import PowerTransformer
from sklearn.preprocessing import StandardScaler
pt=PowerTransformer()
sc=StandardScaler()
X_transform=pt.fit_transform(X)
X=pd.DataFrame(X_transform,columns=X.columns)
X_scale=sc.fit_transform(X)
X=pd.DataFrame(X_scale,columns=X.columns)

test_df_transform=pt.fit_transform(test_df)
test_df=pd.DataFrame(test_df_transform,columns=test_df.columns)
test_df_scale=sc.fit_transform(test_df)
test_df=pd.DataFrame(test_df_scale,columns=test_df.columns)
```

```python
X
```

```python
from lightgbm import LGBMRegressor
light=LGBMRegressor()
light.fit(X,y)
```

```python
y_pred=light.predict(test_df)
y_pred
```
