```python
import tensorflow
from tensorflow.keras.preprocessing.text import one_hot
```

```python
sent=[ 'the glass of milk',
     'the glass of juice',
     'the cup of tea',
    'I am a good boy',
     'I want to be a tensorflow developer',
     'understand the meaning of words',
     'your videos are amazing']
```

```python
print(sent)
```

```python
voc_size=10000
one_hot_rep=[one_hot(words,voc_size) for words in sent]
one_hot_rep
```

```python
from tensorflow.keras import models,layers
from tensorflow.keras.preprocessing import sequence
import numpy as np
```

```python
sent_length=10
embedded_docs=sequence.pad_sequences(one_hot_rep,padding='post',maxlen=sent_length)
embedded_docs
```

```python
dim=9
nn=models.Sequential([
    
    layers.Embedding(voc_size,9,input_length=sent_length)
    
    
])
```

```python
nn.compile(optimizer='adam',loss='binary_crossentropy',metrics='accuracy')
```

```python
nn.summary()
```

```python
nn.predict(embedded_docs)
```
