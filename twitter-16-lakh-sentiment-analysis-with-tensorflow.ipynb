```python
import numpy as np
import pandas as pd
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing import sequence
from tensorflow.keras import models,layers
from sklearn.model_selection import train_test_split
data=pd.read_csv('../input/sentiment140/training.1600000.processed.noemoticon.csv',encoding='latin1',header=None)
data.head()
```

```python
data=data.drop(columns=[1,2,3,4],axis=1)
data
```

```python
data.shape
```

```python
data.isnull().sum()
```

```python
data.columns=['sentiment','tweets']
data.head()
```

```python
data['sentiment'].value_counts()
```

```python
data1=data[data['sentiment']==0].sample(2500)
data2=data[data['sentiment']==4].sample(2500)
```

```python
df=data1.append(data2)
```

```python
df.shape
```

```python
text=df['tweets'].to_list()
```

```python
text
```

```python
token=Tokenizer()
token.fit_on_texts(text)
```

```python
len(token.index_word)
```

```python
vocab_size=len(token.word_index)+1
vocab_size
```

```python
encoded_text=token.texts_to_sequences(text)
print(encoded_text)
```

```python
max_length=200
x=sequence.pad_sequences(encoded_text,maxlen=max_length,padding='pre')
x
```

```python
x.shape
```

```python
y=df['sentiment']
y.head()
```

```python
y.value_counts()
```

```python
y=y.map({0:0,4:1})
y.shape
```

```python
y.value_counts()
```

```python
vec_size=500
nn=models.Sequential([
    
    layers.Embedding(vocab_size,vec_size,input_length=max_length),
    layers.Conv1D(64,8,activation='relu'),
    layers.MaxPooling1D(2),
    layers.Dropout(0.8),
    
    layers.Dense(40,activation='relu'),
    layers.Dropout(0.9),
    
    layers.Dense(29,activation='relu'),
    layers.GlobalMaxPooling1D(),
    
    layers.Dense(1,activation='sigmoid')
    
    
    ])
```

```python
nn.summary()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20)
nn.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
nn.fit(x_train,y_train,validation_data=(x_test,y_test),epochs=25)
```

```python
nn.evaluate(x_train,y_train)
```

```python
nn.evaluate(x_test,y_test)
```

```python
def get_encoded(x):
    x=token.texts_to_sequences(x)
    x=sequence.pad_sequences(x,maxlen=max_length,padding='pre')
    return x
```

```python
h=get_encoded(h)
h
```
