# Lets start the clustering work

```python
import numpy as np
np.arange(1,100)
```

```python
from sklearn.cluster import KMeans
```

```python
sse=[]

k_rng=np.arange(1,25)
for k in k_rng:
    km=KMeans(n_clusters=k)
    km.fit(data)
    sse.append(km.inertia_)
```

```python
sse
```

```python
plt.plot(k_rng,sse)
plt.xlabel('Value of K')
plt.ylabel('Sum of squared error')
plt.show()
```

```python
km=KMeans(n_clusters=5)
km.fit(data)
```

```python
y_pred=km.predict(data)
y_pred
```

```python
data['cluster']=y_pred
data.head()
```

```python
df=data.copy()
```

```python
km.cluster_centers_
```

```python
main_data=data.drop(columns='cluster',axis=1)
df1=df[df.cluster==0]
df2=df[df.cluster==1]
df3=df[df.cluster==2]
df4=df[df.cluster==3]
df5=df[df.cluster==4]
plt.figure(figsize=(25,25))
plt.scatter(df1.f_00,df1.f_28,color='green')
plt.scatter(df2.f_00,df2.f_28,color='red')
plt.scatter(df3.f_00,df3.f_28,color='yellow')
plt.scatter(df4.f_00,df4.f_28,color='blue')
plt.scatter(df5.f_00,df5.f_28,color='purple')
plt.legend()
plt.show()
```
