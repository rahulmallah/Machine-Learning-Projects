```python
train_df=train_df.drop(columns='pk',axis=1)
train_df.head()
```

```python
x=train_df.drop(columns='target',axis=1)
y=train_df['target']
```

```python
cat_features=[0]
```

```python
from catboost import CatBoostClassifier
cat=CatBoostClassifier(cat_features=cat_features)
```

```python
iterations=[500]
learning_rate=[0.01,0.03,0.001]
depth=[int(x) for x in np.linspace(4,12,3)]
l2_leaf_reg=[3,6]
loss_function=['LogLoss','CrossEntropy']
```

```python
grid={
    
    'iterations':iterations,
    'learning_rate':learning_rate,
    'depth':depth,
    'l2_leaf_reg':l2_leaf_reg,
    'loss_function':loss_function
    
    }
```

```python
from sklearn.model_selection import GridSearchCV
cat_grid=GridSearchCV(estimator=cat,param_grid=grid)
```

```python
cat_grid.fit(x,y)
```

```python
cat_grid.best_score_
```

```python
cat_grid.best_params_
```

```python
best_grid=cat_grid.best_estimator_
best_grid
```
