```python
sample_df=pd.read_csv('../input/iit-g-hackathon/sample.csv')
sample_df.head()
```

```python
df=pd.concat([train_df,val_df],ignore_index=True)
df.head()
```

```python
df['ShapeFactor']=df['ShapeFactor1']+df['ShapeFactor2']+df['ShapeFactor3']+df['ShapeFactor4']
```

```python
df=df.drop(columns=['ShapeFactor1','ShapeFactor2','ShapeFactor3','ShapeFactor4','id'],axis=1)
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['Class']=le.fit_transform(df['Class'])
```

```python
import seaborn as sns
from matplotlib import pyplot as plt
plt.figure(figsize=(25,25))
sns.heatmap(df.corr(),annot=True,cmap='BuPu')
plt.show()
```

```python
import warnings
warnings.filterwarnings('ignore')
plt.figure(figsize=(25,25))
for i, var in enumerate(df.columns):
    plt.subplot(7,2,i+1)
    sns.distplot(df[var],color='blue')
```

```python
x=df.drop(columns='Class',axis=1)
y=df['Class']
x
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
rfg=RandomForestClassifier()
```

```python
test_df['ShapeFactor']=test_df['ShapeFactor1']+test_df['ShapeFactor2']+test_df['ShapeFactor3']+test_df['ShapeFactor4']
```

```python
test_df=test_df.drop(columns=['ShapeFactor1','ShapeFactor2','ShapeFactor3','ShapeFactor4','id'],axis=1)
test_df.head()
```

```python
rfg.fit(x,y)
```

```python
y_pred3=rfg.predict(test_df)
```

```python
submission=pd.DataFrame({'Id':sample_df['Id'],'Predicted':y_pred3})
submission.head()
```

```python
submission['Predicted']=submission['Predicted'].map({0:'BARBUNYA',1:'BOMBAY',2:'CALI',3:'DERMASON',4:'HOROZ',5:'SEKER',6:'SIRA'})
```
